```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("./AB_NYC_2019 (1).csv")
```

```python
df.head()
```

# 1. vẽ biểu đồ cột cho neighborhood group

```python
df.info()
```

```python
value_counts=df['neighbourhood_group'].value_counts()
value_counts
```

```python
# Vẽ biểu đồ cột
plt.bar(value_counts.index, value_counts.values)
plt.title("Neighborhood Group Counts")
plt.xlabel('Neighbourhood Group')
plt.ylabel('Count')
plt.show()
```

# 2. tạo histogram cho neighbourhood

```python
neighborhood = df['neighbourhood']
plt.hist(neighborhood, bins=5)
# Định dạng biểu đồ
plt.xlabel('neighbourhood')
plt.ylabel('Frequency')
plt.title('Neighbourhood Group Histogram')

# Hiển thị biểu đồ
plt.show()
```

# 3. Biểu diễn mối liên hệ giữa neighbourgroup và Availability của các phòng

```python
sns.boxplot(x='neighbourhood_group', y='availability_365', data=df)
sns.set(rc={'figure.figsize':(10,6)})
sns.set_style('whitegrid')
plt.title('Availability by Neighbourhood Group')


plt.show()
```

```python
sns.scatterplot(x='longitude', y='latitude', data=df)
sns.set(rc={'figure.figsize':(15,7)})
sns.set_style('bla')
plt.title('Neighbourhood Map')
plt.show()
```

```python
corr = df.corr()
sns.heatmap(corr, annot=True)
sns.set_style('whitegrid')
plt.title('Correlation Heatmap')
plt.show()
```
